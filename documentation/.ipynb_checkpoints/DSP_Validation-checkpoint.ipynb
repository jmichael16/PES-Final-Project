#### DSP and FFT Validation
This Jupyter Notebook is intended to help verify and compare the results of the ARM CMSIS DSP library when calculating an FFT on a known input dataset. I used Python to generate the C array that I use for the Hanning window in the C program (to make the ends of the input data seem continuous). Then, I will generated a scaled sinusoid with various known frequency components. This data will be scaled properly to a `uint16_t` datatype and copied into the test dataset in my C test function. Note that my implementation in C of the the ARM CMSIS library will use the fixed-point Q1.15 datatype `q15_t`. Therefore, the analysis in this notebook uses various scaling factors to make the comparison.   

```python
# use below libraries:
import math 
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from numpy import genfromtxt

# constants, defined consistently with the C program:
# the C code will use Q1.15 datatype - 1 sign bit and 15 fractional bits
N = 256
fs = 48000
```

```python
# Generate the Hanning window:
window = np.hanning(N)
idx = np.linspace(0, N-1, N)
plt.plot(idx, window)
plt.title("window")
plt.show()
```

```python
# generate the samples input for C test
freqs = [1000, 2000, 5000, 10000, 15000]
t = np.linspace(0, N/fs, N)
y = np.zeros(len(t))

# sum the sinusoidal components at those freqs
for i in range(len(freqs)):
    y = y+np.sin(2*math.pi*freqs[i]*t)

nrmlz = round(max(abs(min(y)),abs(max(y))))
y = (y*2**15/nrmlz)
# uncomment following line when generating data for C input dataset
#y = np.round(y-min(y))
print("max: ", max(y), ", min: ", min(y), sep="")

# plot the waveform
plt.plot(t, y, label="combined sinusoid")
plt.plot(t, y*window, label="windowed sinusoid")
plt.title("N={} samples at fs={}".format(len(t),fs))
plt.legend()
plt.show()
for i in range(len(y)):
    print(round(y[i]*window[i]))
```

```python
# do an fft to extract freqs:
cmplx = np.fft.rfft(window*y)
mag = np.abs(cmplx)
freq_axis = np.fft.rfftfreq(y.size, d=1./fs)
# find peaks:
peaks, _ = find_peaks(mag, height=100000)
plt.plot(mag)
plt.plot(peaks, mag[peaks], "x")
plt.show()

# print indices of peaks:
print("idx |  peak freq")
print("----------------")
for i in range(len(peaks)):
    print("{:>3} | {:>10}".format(peaks[i], int(freq_axis[peaks[i]])) )

print("\n\nidx |   freq     | mag  ")
print("-------------------------")
for i in range(len(mag)):
    print("{:>3} | {:>10} | {:>10}".format(i, int(freq_axis[i]), int(mag[i])))
```

```python
# print window to C array
# shift upward to put in uint16_t range
y = np.round(y-min(y))
print("static const int16_t window[NSAMPLES] = {")
for i in range(len(window)):
    print("{:>5}".format(int(window[i]*2**15)), end="")
    if (i!=len(window)-1):
        print(", ", end="")
    else:
        print("};", end="")
    if (i%8==7):
        print("\n", end="")

# print normalized (uint16_t) samples to C array for test waveform  
print("\n\nuint16_t samples_in[NSAMPLES] = {")
for i in range(len(y)):
    print("{:>6}".format(int(y[i])), end="")
    if (i!=len(window)-1):
        print(", ", end="")
    else:
        print("};", end="")
    if (i%8==7):
        print("")
```

```python
f = open("fft_mag_C_output.txt")
lines = f.readlines()
freqs = []
for i,line in enumerate(lines):
    temp = line.strip(" \r\n").split(",")
    for j in range(len(temp)):
        if temp[j] != '':
            freqs.append(int(temp[j]))

"""
translation of following C algorithm in Python

// algorithm only works for stride = 128/8 = 16 
  int stride = 16;
  int16_t peak = 0;
  int peak_idx = 0;

  for (int i=0; i<FFT_MAG_LEN; i+=stride) {
    peak = fft_mag[i];
    for (int j=0; j<stride; j++) {
      if (peak <= fft_mag[i+j]) {
        peak = fft_mag[i+j];
        peak_idx = i+j;
      }
    }
    dest->mags[i>>4] = peak;
    dest->indices[i>>4] = peak_idx;
"""

peak = np.zeros(8)
bins = np.zeros(8)
i=0
tmp_peak = 0
tmp_peak_idx = 0

while i<128:
    tmp_peak = freqs[i]
    for j in range(16):
        if tmp_peak <= freqs[i+j]:
            tmp_peak = freqs[i+j]
            tmp_peak_idx = i+j
    
    peak[i>>4] = tmp_peak
    bins[i>>4] = tmp_peak_idx
    i+=16
    
for i in range(len(bins)):
    print("{:d},".format(int(bins[i])),end="")
print("")
for i in range(len(peak)):
    print("{:d},".format(int(peak[i])),end="")
```
